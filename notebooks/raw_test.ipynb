```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os, sys
```

```python
from_dir='/usr/sci/cedmav/files/raw_files/'
```

```python
def visualize(data,cmap=None):
    fig=plt.figure(figsize = (15,15))        
    plot = plt.imshow(data, origin='lower', cmap=cmap, aspect='auto')
    plt.show()
    return plot
    
    
```

```python
a=[]
for files in os.listdir(from_dir):
    a.append(files)
a.sort()
path=os.path.join(from_dir,'lev_50_face_0.raw')
# path
with open(path,'rb') as f:
    cont=f.read()
    
    data=np.frombuffer(cont, dtype='float32')
    data=data.reshape(1024,1440,1440)
    visualize(data[16,:,:],cmap='coolwarm')
    
```
